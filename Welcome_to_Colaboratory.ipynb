```
import numpy as np
import pandas as pd
```

```
import matplotlib.pyplot as plt 
```

```
data=pd.read_csv('/content/diabetes.csv')
```

```
data.head(10)
```

```
np.mean(data['Pregnancies'])
```

```
np.mean(data["BMI"])
```

```
np.mean(data['BloodPressure'])
```

```
np.mean(data['Glucose'])
```

```
plt.plot(data['Age'],data['Pregnancies'])
plt.title("Age vs Pregnancies")
plt.xlabel("Age")
plt.ylabel("Pregnancy")
plt.show()
```

```
plt.plot(data['Age'],data['Pregnancies'],"go")
plt.title("Age vs Pregnancies")
plt.xlabel("Age")
plt.ylabel("Pregnancy")
plt.show()
```

```
plt.plot(data['Age'],data['BMI'])
plt.title("Age vs BMI")
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()
```

```
divisions=data['Outcome']
division_scores=data['Age']
plt.bar(divisions,division_scores,color='blue')
plt.title("Bar Graph")
plt.xlabel("Outcomes")
plt.ylabel("Age")
plt.show()
```
